```python
from random import uniform

def goal(x):
    return x*x*(x - 1)*(x - 2)*(x -3)*(x - 4)


class State:
    
    def __init__(self, x):
        self.x = x
        self.f = goal(x)


def sucessor(s):
    x1 = s.x  + 0.001
    vizinho1 = State(x1)
    x2 = s.x - 0.001
    vizinho2 = State(x2)
    if(vizinho1.f < vizinho2.f):
        return vizinho1
    return vizinho2

def initialState():
    return State(uniform(-10, 10))


melhor = initialState()
i = 0

while i < 100:
    atual = initialState()
    while(True):
        vizinho = sucessor(atual)
        if(vizinho.f >= atual.f):
            break
        atual = vizinho
    if(atual.f < melhor.f):
        melhor = atual

    i += 1

print(f'x = {melhor.x}, f = {melhor.f}')
```
